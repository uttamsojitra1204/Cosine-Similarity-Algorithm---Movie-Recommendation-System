```python
#importing the neccesary library
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#read the dataset
df= pd.read_csv("movies.csv")
```

```python
#print first 5 rows
df.head(5)
```

```python
#how many rows and columns are there
df.shape
```

```python
#printing columns name
df.columns
```

```python
#select the relevant feature for movie recommendation
selected_features= ["genres", "keywords", "tagline", "cast", "director"]
print(selected_features)
```

```python
#replacing the null value with null string
for feature in selected_features:
    df[feature]= df[feature].fillna("")
```

```python
#combining all the 5 selected features
combined_features= df["genres"] + df["keywords"] + df["tagline"] + df["cast"] + df["director"]
```

```python
print(combined_features)
```

```python
#converting the text data into feature vectorizer
vectorizer= TfidfVectorizer()
```

```python
feature_vector= vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

```python
#cosine similarity
```

```python
#getting the similarity scores using cosine similarity
similarity= cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
#getting the movie name from the user
movie_name= input("Enter your favorite movie name: ")
```

```python
#creating the list all the movie name given in the dataset
list_of_movie= df["title"].tolist()
print(list_of_movie)
```

```python
#finding the close match for the movie name given by user
find_close_match= difflib.get_close_matches(movie_name, list_of_movie)
print(find_close_match)
```

```python
close_match= find_close_match[0]
print(close_match)
```

```python
#find the index with movie with title
movie_index= df[df.title == close_match]["index"].values[0]
print(movie_index)
```

```python
#getting a list of similar movie
similarity_score= list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movie score based on their score
sorted_similarity_score= sorted(similarity_score, key= lambda x: x[1], reverse= True)
print(sorted_similarity_score)
```

```python
#print the similar movie based on the index
print("movie suggested for you: \n")
i= 1
for movie in sorted_similarity_score:
    index= movie[0]
    title_from_index= df[df.index== index]["title"].values[0]
    if i<30:
        print(i, " ", title_from_index)
        i+=1
```

```python
movie_name= input("enter the movie name: ")

movie_list= df["title"].tolist()

close_match= difflib.get_close_matches(movie_name, movie_list)

close_match= close_match[0]

movie_index= df[df.title == close_match]["index"].values[0]

similarity_score= list(enumerate(similarity[movie_index]))

sorted_similarity_score= sorted(similarity_score, key= lambda x: x[1], reverse= True)

print("movie suggested for you : \n")

i= 1
for movie in sorted_similarity_score:
    index= movie[0]
    title_from_index= df[df.index == index]["title"].values[0]
    if i < 30:
        print(i, "." ,title_from_index)
        i += 1
```
